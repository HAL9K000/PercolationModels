```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import matplotlib
from matplotlib import colors
```

```python
fin_1 = open('np_frame_512_0.45.txt','r')
fin_2 = open('dp_frame_512_0.692.txt','r')
fin_3 = open('tp_frame_512_0.8_0.38.txt','r')

a_1=[]
a_2=[]
a_3=[]

for line in fin_1.readlines():
    a_1.append( [ int (x) for x in line[0:-1]] )
    
for line in fin_2.readlines():
    a_2.append( [ int (x) for x in line[0:-1]] )
    
for line in fin_3.readlines():
    a_3.append( [ int (x) for x in line[0:-1]] )
```

```python
b_1 = np.array(a_1)
b_2 = np.array(a_2)
b_3 = np.array(a_3)
```

```python
from matplotlib.colors import LinearSegmentedColormap
cMap = []
for value, colour in zip([0,1],["palegoldenrod", "forestgreen"]):
    cMap.append((value/1., colour))

customColourMap = LinearSegmentedColormap.from_list("custom", cMap)
plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.imshow(b_1, animated=True,cmap=customColourMap)
plt.axis('off')
plt.title('P (p=0.45)',fontsize=20)
plt.subplot(2,3,2)
plt.imshow(b_2, animated=True,cmap=customColourMap)
plt.axis('off')
plt.title('DP (p=0.692)',fontsize=20)
plt.subplot(2,3,3)
plt.imshow(b_3, animated=True,cmap=customColourMap)
plt.axis('off')
plt.title('TP (p=0.38, q=0.8)',fontsize=20)

plt.subplot(2,3,4)
plt.imshow(b_1[0:int(512/4),0:int(512/4)], animated=True,cmap=customColourMap)
plt.axis('off')
plt.title('4x Magnification',fontsize=20)
plt.subplot(2,3,5)
plt.imshow(b_2[0:int(512/4),0:int(512/4)], animated=True,cmap=customColourMap)
plt.axis('off')
plt.title('4x Magnification',fontsize=20)
plt.subplot(2,3,6)
plt.imshow(b_3[0:int(512/4),0:int(512/4)], animated=True,cmap=customColourMap)
plt.axis('off')
plt.title('4x Magnification',fontsize=20)

plt.tight_layout()
#plt.savefig("snapshot", dpi = 100, bbox_inches='tight')
```
