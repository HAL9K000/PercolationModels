```python
import pickle
import numpy as np
import collections
import matplotlib.pyplot as plt
import copy
import matplotlib.ticker as ticker
```

```python
density = 0.56
grid_size = 1024
replicates = [1,2,3,4,5]
numbers_per_replicate = 2000000
dpt = []
npt = []
for replicate in replicates:
    filename_dp = "dp_transformations_" + str(grid_size) + "_" + str(density) + "_" + str(numbers_per_replicate) + "_" + str(replicate) + ".txt"
    with open(filename_dp) as f:
        dpt.extend([tuple(map(int, i.split(' '))) for i in f])
```

```python
var_dpsc = [(i[0],i[1]-i[0]) for i in dpt]
var_dpsc_p = [i for i in var_dpsc if i[1]>=0]
var_dpsc_n = [i for i in var_dpsc if i[1]<0]

x_dp_p = [i[0] for i in var_dpsc_p]
x_dp_n = [i[0] for i in var_dpsc_n]
y_dp_p = [i[1] for i in var_dpsc_p]
y_dp_n = [i[1] for i in var_dpsc_n]
y2_dp_p = [i[1]**2 for i in var_dpsc_p]
y2_dp_n = [i[1]**2 for i in var_dpsc_n]
```

```python
fig, ax1 = plt.subplots(figsize=(5,5))

ax1.scatter(x_dp_p,y_dp_p,s=0.05,c='g',alpha=0.6)
ax1.scatter(x_dp_n,y_dp_n,s=0.05,c='r',alpha=0.6)
ax1.ticklabel_format(style='sci', axis='x', scilimits=(0,0),useMathText=True)
ax1.ticklabel_format(style='sci', axis='y', scilimits=(0,0),useMathText=True)
ax1.set_xlabel('$s$',fontsize=20)
ax1.set_ylabel('$\Delta s$',fontsize=20)
ax1.set_title('$DP: \\rho = $'+str(density))

#plt.savefig("Figure_7",dpi=300)
```
