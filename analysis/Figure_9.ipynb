```python
data_32 = []

filename_np = "average_cluster_size_32.txt"
with open(filename_np) as f:
    data_32.extend([tuple(map(float, i.split(' '))) for i in f])  
    
np_densities_32 = [i[0] for i in data_32]
np_ps_32 = np.linspace(0.2,0.8,200)
dp_densities_32 = [i[2] for i in data_32]
dp_ps_32 = np.linspace(0.638,0.843,200)

np_avg_cluster_sizes_32 = [i[1] for i in data_32]
dp_avg_cluster_sizes_32 = [i[3] for i in data_32]
```

```python
data_256 = []

filename_np = "average_cluster_size_256.txt"
with open(filename_np) as f:
    data_256.extend([tuple(map(float, i.split(' '))) for i in f]) 
    
np_densities_256 = [i[0] for i in data_256]
np_ps_256 = np.linspace(0.2,0.8,400)
dp_densities_256 = [i[2] for i in data_256]
dp_ps_256 = np.linspace(0.638,0.843,400)

np_avg_cluster_sizes_256 = [i[1] for i in data_256]
dp_avg_cluster_sizes_256 = [i[3] for i in data_256]
```

```python
fig, ax1 = plt.subplots(figsize=(5,4))

left_dp, bottom_dp, width_dp, height_dp = [0.18, 0.6, 0.22, 0.22]
ax2 = fig.add_axes([left_dp, bottom_dp, width_dp, height_dp])

left_np, bottom_np, width_np, height_np = [0.67, 0.29, 0.22, 0.20]
ax3 = fig.add_axes([left_np, bottom_np, width_np, height_np])


ax1.plot(np_densities_256,np_avg_cluster_sizes_256,label='P',color='black')
ax1.plot(dp_densities_256,dp_avg_cluster_sizes_256,label='DP',color='red')
ax1.set_xlim(0.15,0.9)
ax1.set_ylim(0.0,25.0)
ax1.set_xlabel('$\\rho_{eq}$ (Density)',fontsize=15)
ax1.set_ylabel('$S$ (average cluster size)',fontsize=15)
ax1.legend()


ax2.plot(dp_ps_32,dp_avg_cluster_sizes_32,color='magenta')
ax2.plot(dp_ps_64,dp_avg_cluster_sizes_64,color='green')
ax2.plot(dp_ps_128,dp_avg_cluster_sizes_128,color='blue')
ax2.plot(dp_ps_256,dp_avg_cluster_sizes_256,color='orange')
ax2.set_xlabel('$p$',fontsize=8)
ax2.tick_params(labelsize=8)
ax2.set_xticks([0.728])
ax2.axvline(x=0.728,color='black')
ax2.text(0.2, 0.8, 'DP', transform=ax2.transAxes, ha="right")

ax3.plot(np_ps_32,np_avg_cluster_sizes_32,color='magenta',label='32')
ax3.plot(np_ps_64,np_avg_cluster_sizes_64,color='green',label='64')
ax3.plot(np_ps_128,np_avg_cluster_sizes_128,color='blue',label='128')
ax3.plot(np_ps_256,np_avg_cluster_sizes_256,color='orange',label='256')
ax3.set_xlabel('$p$',fontsize=8)
ax3.tick_params(labelsize=8)
ax3.set_xticks([0.592])
ax3.set_yticks([10,20])
ax3.axvline(x=0.592,color='black')
ax3.text(0.12, 0.8, 'P', transform=ax3.transAxes, ha="right")
ax3.legend(bbox_to_anchor=(1.05, 2.3))

plt.savefig("Figure_9_300.png",dpi=300)
```
