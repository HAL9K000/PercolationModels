```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_file(filename):
    ps = [] 
    densities = []
    g_32 = []
    g_64 = []
    g_128 = []
    g_256 = []
    g_512 = []
    infile = open(filename, "r") 
    for line in infile: 
        numbers = line.split() 
        densities_ = float(numbers[0]) 
        ps_ = float(numbers[1])
        g_32_ = float(numbers[2])
        g_64_ = float(numbers[3]) 
        g_128_ = float(numbers[4])
        g_256_ = float(numbers[5])
        g_512_ = float(numbers[6])
        
        densities.append(densities_)
        ps.append(ps_)
        g_32.append(g_32_)
        g_64.append(g_64_)
        g_128.append(g_128_)
        g_256.append(g_256_)
        g_512.append(g_512_)
        
    return densities, ps, g_32, g_64, g_128, g_256, g_512
    return densities, ps, g_32, g_64, g_128, g_256, g_512
```

```python
np_densities, np_ps, np_P_32, np_P_64, np_P_128, np_P_256, np_P_512 =read_file("np_new.txt") 
dp_densities, dp_ps, dp_P_32, dp_P_64, dp_P_128, dp_P_256, dp_P_512 =read_file("dp_new.txt") 
```

```python
fig, ax1 = plt.subplots(figsize=(5,4))

left_np, bottom_np, width_np, height_np = [0.18, 0.6, 0.22, 0.18]
ax2 = fig.add_axes([left_np, bottom_np, width_np, height_np])

left_dp, bottom_dp, width_dp, height_dp = [0.65, 0.25, 0.22, 0.18]
ax3 = fig.add_axes([left_dp, bottom_dp, width_dp, height_dp])


ax1.scatter(np_densities,np_P_512,s=20,marker='d',label='P',color='black')
ax1.scatter(dp_densities,dp_P_512,s=20,marker='o',label='DP',color='red')
ax1.set_xlim(0.5,0.66)
ax1.set_ylim(0.0,1.0)
ax1.set_xlabel('$\\rho$ (density)',fontsize=13)
ax1.set_ylabel('probability of percolation',fontsize=13)
ax1.set_xticks([0.561,0.592])
ax1.set_yticks([0,1])
ax1.axvline(x=0.561,color='black',linestyle='--')
ax1.axvline(x=0.592,color='black',linestyle='--')
ax1.legend()

ax2.plot(dp_ps,dp_P_512,color='green')
ax2.plot(dp_ps,dp_P_256,color='red')
ax2.plot(dp_ps,dp_P_128,color='blue')
ax2.plot(dp_ps,dp_P_64,color='black')
ax2.plot(dp_ps,dp_P_32,color='orange')
ax2.set_xlim(0.71,0.75)
ax2.set_xlabel('$p$',fontsize=8)
ax2.tick_params(bottom=True,top=True,left=True,right=True)
ax2.tick_params(labelsize=8)
ax2.set_xticks([0.728])
ax2.set_yticks([0,1])
ax2.axvline(x=0.728,color='black')
ax2.text(0.2, 0.8, 'DP', transform=ax2.transAxes, ha="right")

ax3.plot(np_ps,np_P_512,color='green',label='512')
ax3.plot(np_ps,np_P_256,color='red',label='256')
ax3.plot(np_ps,np_P_128,color='blue',label='128')
ax3.plot(np_ps,np_P_64,color='black',label='64')
ax3.plot(np_ps,np_P_32,color='orange',label='32')
ax3.set_xlim(0.56,0.63)
ax3.set_xlabel('$p$',fontsize=8)
ax3.tick_params(bottom=True,top=True,left=True,right=True)
ax3.tick_params(labelsize=8)
ax3.set_xticks([0.592])
ax3.set_yticks([0,1])
ax3.axvline(x=0.592,color='black')
ax3.text(0.1, 0.8, 'P', transform=ax3.transAxes, ha="right")
ax3.legend(bbox_to_anchor=(0.3, 1.1))
#plt.savefig("Figure_1.png",dpi=1000)
```
